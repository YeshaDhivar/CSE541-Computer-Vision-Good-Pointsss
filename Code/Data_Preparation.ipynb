```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
Raw_DIR= r'E:\Yesha\3rd Year\6th Sem\CV\Project\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        #0 in the file name denotes close eye 
        #1 in the file name denotes open eye
        if i.split('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'E:\Yesha\3rd Year\6th Sem\CV\Project\mrlEyes_2018_01\Prepared Data\Close Eyes')
        
        elif i.split('_')[4]=='1':
            shutil.copy(src=dirpath+'/'+i, dst=r'E:\Yesha\3rd Year\6th Sem\CV\Project\mrlEyes_2018_01\Prepared Data\Open Eyes')
```
